The advanced statistics will be used to predict playoff win percentage using the XGBoost Regressor.

```python
X = np.array(dataset.iloc[:,1:26])
Y = np.array(dataset.iloc[:,27:28])
Z = dataset.iloc[:,1:26]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=40)
```

```python
model = XGBRegressor(subsample=0.6, n_estimators=500, max_depth=3, learning_rate=0.01, colsample_bytree=0.7, colsample_bylevel=0.6)
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
r2_score(y_pred, y_test)
```

```python
predictions = [round(value,2) for value in y_pred]
print(predictions)
print(mean_squared_error(y_test,predictions))
```

The code below randomly chooses a stat for each team, and if one team is higher in that stat than the other team, they get a point. This is not an accurate way to see which team is better because sometimes you want a stat to be lower than the other team, such as TOV%.

```python
import random
one = metadata.loc[Team1]
two = metadata.loc[Team2]

def gamesim(number_simulations):
  moment = 0
  z = int(random.choice(np.linspace(3,24, num=22)))
  for i in range(0,number_simulations):
    if float(one[z]) > float(two[z]):
      moment +=1
    else:
     moment -=1
  if moment > 0:
    print("The winner is likely", Team1)
  else:
    print("The winner is likely", Team2)
  

gamesim(50)
```
